```python
new_method = [9.64, 15.43, 119.33]
```

```python
def avg(l1, l2):
    return [(a + b) / 2 for a, b in zip(l1, l2)]
```

```python
labels = ['20_5', '50_10', '100_10']
methods = ['NEH', 'Greedy NEH', 'Palmer', 'CDS', 'PRSKE', 'Artificial', 'IG-BS',]
```

```python
data = [
    avg(neh1, neh2),
    avg(greedy_neh1, greedy_neh2),
    avg(palmer1, palmer2),
    avg(cds1, cds2),
    avg(PRSKE1, PRSKE2),
    avg(artificial1, artificial2), 
    new_method
]
```

```python
x = np.arange(len(labels))  
width = 0.13 

fig, ax = plt.subplots(figsize=(10, 6), facecolor='none')
ig, ax = plt.subplots(figsize=(10, 6), facecolor='none')
ax.set_facecolor('none')
fig.patch.set_alpha(0)

for i in range(len(data)):
    ax.bar(x + i*width, data[i], width, label=methods[i])

ax.set_xlabel('Problmes')
ax.set_ylabel('Temps de calcul (s)')
ax.set_title('Comparaison des mthodes en temps de calcul')
ax.set_xticks(x + width*2.5)
ax.set_xticklabels(labels)
ax.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
